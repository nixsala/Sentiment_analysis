```python
from transformers import pipeline
```

```python
model=pipeline('sentiment-analysis')
```

```python
paragraph="I had an bore day today!.The weather was gloomy and I had a bad time with my friends.I did not get  a promotion at work!."
```

```python
result=model(paragraph)[0]
```

```python
if result['label']=='POSITIVE':
  print("the paragraph is positive",result['score'])
else:
  print("the paragraph is negative",result['score'])
```
